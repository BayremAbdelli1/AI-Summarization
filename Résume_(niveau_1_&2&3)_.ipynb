```python
from langchain.document_loaders import PyPDFLoader
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains.llm import LLMChain
from langchain_core.prompts import PromptTemplate
loader = PyPDFLoader("D2011.pdf")
pages = loader.load()


from langchain_openai import ChatOpenAI
prompt_template = """Write a concise summary of the following in french:
"{text}"
CONCISE SUMMARY:"""
prompt = PromptTemplate.from_template(prompt_template)
llm = ChatOpenAI(temperature=0, model_name="gpt-4o")
llm_chain = LLMChain(llm=llm, prompt=prompt)
stuff_chain = StuffDocumentsChain(llm_chain=llm_chain, document_variable_name="text")
print(stuff_chain.invoke(pages)["output_text"])
```

```python
# Define the prompt for Niveau 2 summary
prompt_template = """Write a detailed summary of the following in French:
"{text}"

Please include any sanctions or penalties mentioned in the text.

DETAILED SUMMARY WITH SANCTIONS:"""
prompt = PromptTemplate.from_template(prompt_template)
llm = ChatOpenAI(temperature=0, model_name="gpt-4o")
llm_chain = LLMChain(llm=llm, prompt=prompt)
stuff_chain = StuffDocumentsChain(llm_chain=llm_chain, document_variable_name="text")
detailed_summary = stuff_chain.invoke(pages)["output_text"]
print(detailed_summary)
```

```python
# Load the current (modified) PDF document
loader_current = PyPDFLoader("newD2011.pdf")
pages_current = loader_current.load()
# Load the previous version of the do
loader_previous = PyPDFLoader("D2011.pdf")
pages_previous = loader_previous.load()

prompt_template = """Write a detailed summary of the following in French:

Current Version:
"{current_text}"

Include any sanctions or penalties mentioned in the text.

Additionally, compare it with the previous version provided below:

Previous Version:
"{previous_text}"

DETAILED SUMMARY WITH SANCTIONS AND COMPARISON:"""
prompt = PromptTemplate.from_template(prompt_template)
llm = ChatOpenAI(temperature=0, model_name="gpt-4o")
llm_chain = LLMChain(llm=llm, prompt=prompt)
current_text = "\n".join([page.page_content for page in pages_current])
previous_text = "\n".join([page.page_content for page in pages_previous])
input_data = {
    "current_text": current_text,
    "previous_text": previous_text
}
# Generate and print the detailed summary with comparison
detailed_summary_comparison = llm_chain.run(input_data)
print(detailed_summary_comparison)
```
